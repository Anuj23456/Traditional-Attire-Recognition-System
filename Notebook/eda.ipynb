```python
from sklearn.model_selection import train_test_split
```

```python
import pickle
image_data = pickle.load(open('mydata.pkl','rb'))
```

```python
import random
random.shuffle(image_data)
features=[]
labels=[]

for feature,label in image_data:
    features.append(feature)
    labels.append(label)

```

```python
x_train,x_test,y_train,y_test= train_test_split(features,labels,test_size=0.3)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='poly',gamma='auto',decision_function_shape='ovr')
model.fit(x_train,y_train)
```

```python
pickle.dump(model,open('svcmodel.pkl','wb'))
```

```python
prediction = model.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(prediction,y_test))
print(confusion_matrix(prediction,y_test))
```

```python
import matplotlib.pyplot as plt
dress = x_test[45].reshape(150,150)
categories = ['Adi','Apatani','Galo','Mishmi','Monpa','Nocte','Nyishi','Singpho','Tai Khampti']

print("prediction is", categories[prediction[45]])
plt.imshow(dress,cmap='gray')
plt.show()
```
